```python
"""
AI Ethics Radar - Model Training Pipeline (FIXED)
Trains 6 models total:
- Set 1 (Stage 1 - 6 labels): TF-IDF + LogReg, DistilBERT, RoBERTa
- Set 2 (Stage 2 - 12 labels): TF-IDF + LogReg, DistilBERT, RoBERTa

Hardware: Kaggle T4 x2 GPU
"""

import pandas as pd
import numpy as np
import torch
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import f1_score, hamming_loss, classification_report
from transformers import (
    AutoTokenizer, 
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer,
    EarlyStoppingCallback
)
from datasets import Dataset
import time
import matplotlib.pyplot as plt
import seaborn as sns
import json

# ============================================================================
# CONFIGURATION
# ============================================================================

# File paths - UPDATE THESE
TRAIN_STAGE1 = '/kaggle/input/labeled-articles/train_stage1.csv'
VAL_STAGE1 = '/kaggle/input/labeled-articles/val_stage1.csv'
TEST_STAGE1 = '/kaggle/input/labeled-articles/test_stage1.csv'

TRAIN_STAGE2 = '/kaggle/input/labeled-articles/train_stage2.csv'
VAL_STAGE2 = '/kaggle/input/labeled-articles/val_stage2.csv'
TEST_STAGE2 = '/kaggle/input/labeled-articles/test_stage2.csv'

# Output directories
OUTPUT_DIR_STAGE1 = './models_stage1'
OUTPUT_DIR_STAGE2 = './models_stage2'

# Model configurations
MAX_LENGTH = 512  # Maximum token length for transformers
TFIDF_MAX_FEATURES = 10000  # Maximum features for TF-IDF
RANDOM_STATE = 42

# Training hyperparameters
BATCH_SIZE = 8  # Adjust based on GPU memory
LEARNING_RATE = 2e-5
NUM_EPOCHS = 3
WARMUP_STEPS = 500
WEIGHT_DECAY = 0.01

# Device
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {DEVICE}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"GPU Memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")

# ============================================================================
# HELPER FUNCTIONS
# ============================================================================

def load_data(train_path, val_path, test_path):
    """Load train/val/test datasets"""
    print(f"Loading data from:")
    print(f"  Train: {train_path}")
    print(f"  Val: {val_path}")
    print(f"  Test: {test_path}")
    
    train_df = pd.read_csv(train_path)
    val_df = pd.read_csv(val_path)
    test_df = pd.read_csv(test_path)
    
    print(f"\nDataset sizes:")
    print(f"  Train: {len(train_df)}")
    print(f"  Val: {len(val_df)}")
    print(f"  Test: {len(test_df)}")
    
    return train_df, val_df, test_df

def prepare_labels(df, label_prefix):
    """Extract label columns and prepare multi-label arrays"""
    label_cols = [col for col in df.columns if col.startswith(label_prefix)]
    label_cols = sorted(label_cols)  # Ensure consistent ordering
    
    print(f"\nFound {len(label_cols)} labels:")
    for i, col in enumerate(label_cols, 1):
        label_name = col.replace(label_prefix, '').replace('_', ' ').title()
        positive_count = df[col].sum()
        print(f"  {i}. {label_name}: {positive_count} positive samples")
    
    X = df['content'].values
    y = df[label_cols].values
    
    return X, y, label_cols

def compute_metrics(y_true, y_pred):
    """Compute evaluation metrics"""
    micro_f1 = f1_score(y_true, y_pred, average='micro', zero_division=0)
    macro_f1 = f1_score(y_true, y_pred, average='macro', zero_division=0)
    hamming = hamming_loss(y_true, y_pred)
    
    return {
        'micro_f1': micro_f1,
        'macro_f1': macro_f1,
        'hamming_loss': hamming
    }

def compute_metrics_transformer(eval_pred):
    """Compute metrics for Hugging Face Trainer"""
    predictions, labels = eval_pred
    predictions = (predictions > 0.5).astype(int)
    
    micro_f1 = f1_score(labels, predictions, average='micro', zero_division=0)
    macro_f1 = f1_score(labels, predictions, average='macro', zero_division=0)
    hamming = hamming_loss(labels, predictions)
    
    return {
        'micro_f1': micro_f1,
        'macro_f1': macro_f1,
        'hamming_loss': hamming
    }

# ============================================================================
# MODEL 1: TF-IDF + LOGISTIC REGRESSION
# ============================================================================

def train_tfidf_logreg(X_train, y_train, X_val, y_val, X_test, y_test, stage_name):
    """Train TF-IDF + Logistic Regression baseline"""
    print("\n" + "="*80)
    print(f"MODEL 1 ({stage_name}): TF-IDF + LOGISTIC REGRESSION")
    print("="*80)
    
    start_time = time.time()
    
    # Vectorize text
    print("\nVectorizing text with TF-IDF...")
    vectorizer = TfidfVectorizer(
        max_features=TFIDF_MAX_FEATURES,
        ngram_range=(1, 2),
        min_df=2,
        max_df=0.95
    )
    
    X_train_vec = vectorizer.fit_transform(X_train)
    X_val_vec = vectorizer.transform(X_val)
    X_test_vec = vectorizer.transform(X_test)
    
    print(f"  Train shape: {X_train_vec.shape}")
    print(f"  Val shape: {X_val_vec.shape}")
    print(f"  Test shape: {X_test_vec.shape}")
    
    # Train model
    print("\nTraining Logistic Regression (One-vs-Rest)...")
    model = OneVsRestClassifier(
        LogisticRegression(
            max_iter=1000,
            random_state=RANDOM_STATE,
            n_jobs=-1
        )
    )
    
    model.fit(X_train_vec, y_train)
    
    training_time = time.time() - start_time
    print(f"✓ Training completed in {training_time/60:.2f} minutes")
    
    # Evaluate
    print("\n--- Evaluation ---")
    
    # Validation set
    y_val_pred = model.predict(X_val_vec)
    val_metrics = compute_metrics(y_val, y_val_pred)
    print(f"Validation Metrics:")
    print(f"  Micro F1: {val_metrics['micro_f1']:.4f}")
    print(f"  Macro F1: {val_metrics['macro_f1']:.4f}")
    print(f"  Hamming Loss: {val_metrics['hamming_loss']:.4f}")
    
    # Test set
    y_test_pred = model.predict(X_test_vec)
    test_metrics = compute_metrics(y_test, y_test_pred)
    print(f"\nTest Metrics:")
    print(f"  Micro F1: {test_metrics['micro_f1']:.4f}")
    print(f"  Macro F1: {test_metrics['macro_f1']:.4f}")
    print(f"  Hamming Loss: {test_metrics['hamming_loss']:.4f}")
    
    return {
        'model': 'TF-IDF + LogisticRegression',
        'training_time': training_time,
        'val_metrics': val_metrics,
        'test_metrics': test_metrics,
        'model_obj': model,
        'vectorizer': vectorizer
    }

# ============================================================================
# MODEL 2 & 3: TRANSFORMER MODELS (DistilBERT & RoBERTa)
# ============================================================================

def prepare_transformer_dataset(X, y, tokenizer):
    """Prepare dataset for transformer models - FIXED for multi-label"""
    # Tokenize
    encodings = tokenizer(
        X.tolist(),
        truncation=True,
        padding='max_length',
        max_length=MAX_LENGTH,
        return_tensors='pt'
    )
    
    # CRITICAL FIX: Convert labels to float for multi-label classification
    labels_float = y.astype(np.float32)
    
    # Create dataset
    dataset = Dataset.from_dict({
        'input_ids': encodings['input_ids'],
        'attention_mask': encodings['attention_mask'],
        'labels': labels_float.tolist()  # Now as float
    })
    
    return dataset

def train_transformer(model_name, X_train, y_train, X_val, y_val, X_test, y_test, 
                     num_labels, stage_name, output_dir):
    """Train a transformer model (DistilBERT or RoBERTa)"""
    print("\n" + "="*80)
    print(f"MODEL ({stage_name}): {model_name.upper()}")
    print("="*80)
    
    start_time = time.time()
    
    # Load tokenizer and model
    print(f"\nLoading {model_name} tokenizer and model...")
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSequenceClassification.from_pretrained(
        model_name,
        num_labels=num_labels,
        problem_type="multi_label_classification"
    )
    
    # Prepare datasets (with fixed label conversion)
    print("Preparing datasets...")
    train_dataset = prepare_transformer_dataset(X_train, y_train, tokenizer)
    val_dataset = prepare_transformer_dataset(X_val, y_val, tokenizer)
    test_dataset = prepare_transformer_dataset(X_test, y_test, tokenizer)
    
    print(f"  Train: {len(train_dataset)} samples")
    print(f"  Val: {len(val_dataset)} samples")
    print(f"  Test: {len(test_dataset)} samples")
    
    # Training arguments
    model_output_dir = f"{output_dir}/{model_name.split('/')[-1]}"
    training_args = TrainingArguments(
        output_dir=model_output_dir,
        num_train_epochs=NUM_EPOCHS,
        per_device_train_batch_size=BATCH_SIZE,
        per_device_eval_batch_size=BATCH_SIZE,
        learning_rate=LEARNING_RATE,
        weight_decay=WEIGHT_DECAY,
        warmup_steps=WARMUP_STEPS,
        logging_dir=f"{model_output_dir}/logs",
        logging_steps=100,
        eval_strategy="epoch",
        save_strategy="epoch",
        load_best_model_at_end=True,
        metric_for_best_model="micro_f1",
        greater_is_better=True,
        save_total_limit=2,
        report_to="none",
        fp16=torch.cuda.is_available(),  # Use mixed precision if GPU available
    )
    
    # Trainer
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=val_dataset,
        compute_metrics=compute_metrics_transformer,
        callbacks=[EarlyStoppingCallback(early_stopping_patience=2)]
    )
    
    # Train
    print(f"\nTraining {model_name}...")
    print(f"  Epochs: {NUM_EPOCHS}")
    print(f"  Batch size: {BATCH_SIZE}")
    print(f"  Learning rate: {LEARNING_RATE}")
    
    trainer.train()
    
    training_time = time.time() - start_time
    print(f"\n✓ Training completed in {training_time/60:.2f} minutes")
    
    # Evaluate on validation set
    print("\n--- Validation Evaluation ---")
    val_results = trainer.evaluate(val_dataset)
    print(f"Micro F1: {val_results['eval_micro_f1']:.4f}")
    print(f"Macro F1: {val_results['eval_macro_f1']:.4f}")
    print(f"Hamming Loss: {val_results['eval_hamming_loss']:.4f}")
    
    # Evaluate on test set
    print("\n--- Test Evaluation ---")
    test_results = trainer.evaluate(test_dataset)
    print(f"Micro F1: {test_results['eval_micro_f1']:.4f}")
    print(f"Macro F1: {test_results['eval_macro_f1']:.4f}")
    print(f"Hamming Loss: {test_results['eval_hamming_loss']:.4f}")
    
    # Save model and tokenizer
    print(f"\nSaving model to {model_output_dir}...")
    trainer.save_model(model_output_dir)
    tokenizer.save_pretrained(model_output_dir)
    
    return {
        'model': model_name,
        'training_time': training_time,
        'val_metrics': {
            'micro_f1': val_results['eval_micro_f1'],
            'macro_f1': val_results['eval_macro_f1'],
            'hamming_loss': val_results['eval_hamming_loss']
        },
        'test_metrics': {
            'micro_f1': test_results['eval_micro_f1'],
            'macro_f1': test_results['eval_macro_f1'],
            'hamming_loss': test_results['eval_hamming_loss']
        },
        'model_path': model_output_dir
    }

# ============================================================================
# TRAINING PIPELINE
# ============================================================================

def train_stage(train_path, val_path, test_path, stage_name, label_prefix, output_dir):
    """Train all 3 models for a given stage"""
    print("\n" + "="*80)
    print(f"TRAINING {stage_name.upper()}")
    print("="*80)
    
    # Load data
    train_df, val_df, test_df = load_data(train_path, val_path, test_path)
    
    # Prepare labels
    X_train, y_train, label_cols = prepare_labels(train_df, label_prefix)
    X_val, y_val, _ = prepare_labels(val_df, label_prefix)
    X_test, y_test, _ = prepare_labels(test_df, label_prefix)
    
    num_labels = len(label_cols)
    
    results = []
    
    # Model 1: TF-IDF + Logistic Regression
    tfidf_result = train_tfidf_logreg(X_train, y_train, X_val, y_val, X_test, y_test, stage_name)
    results.append(tfidf_result)
    
    # Model 2: DistilBERT
    distilbert_result = train_transformer(
        'distilbert-base-uncased',
        X_train, y_train, X_val, y_val, X_test, y_test,
        num_labels, stage_name, output_dir
    )
    results.append(distilbert_result)
    
    # Model 3: RoBERTa
    roberta_result = train_transformer(
        'roberta-base',
        X_train, y_train, X_val, y_val, X_test, y_test,
        num_labels, stage_name, output_dir
    )
    results.append(roberta_result)
    
    return results, label_cols

# ============================================================================
# MAIN EXECUTION
# ============================================================================

print("\n" + "="*80)
print("AI ETHICS RADAR - MODEL TRAINING PIPELINE")
print("="*80)

# ============================================================================
# STAGE 1: MAIN CATEGORIES (6 labels)
# ============================================================================

stage1_results, stage1_labels = train_stage(
    TRAIN_STAGE1, VAL_STAGE1, TEST_STAGE1,
    'Stage 1 (Main Categories)',
    'stage1_binary_',
    OUTPUT_DIR_STAGE1
)

# ============================================================================
# STAGE 2: AI ETHICS ISSUES (12 labels)
# ============================================================================

stage2_results, stage2_labels = train_stage(
    TRAIN_STAGE2, VAL_STAGE2, TEST_STAGE2,
    'Stage 2 (AI Ethics)',
    'stage2_binary_',
    OUTPUT_DIR_STAGE2
)

# ============================================================================
# RESULTS COMPARISON
# ============================================================================

print("\n" + "="*80)
print("RESULTS COMPARISON")
print("="*80)

def print_comparison_table(results, stage_name):
    """Print comparison table for a stage"""
    print(f"\n--- {stage_name} ---")
    print(f"{'Model':<30} {'Train Time':<15} {'Val Micro F1':<15} {'Val Macro F1':<15} {'Test Micro F1':<15} {'Test Macro F1':<15} {'Test Hamming':<15}")
    print("-" * 125)
    
    for result in results:
        model_name = result['model'].split('/')[-1] if '/' in result['model'] else result['model']
        print(f"{model_name:<30} "
              f"{result['training_time']/60:>10.2f} min  "
              f"{result['val_metrics']['micro_f1']:>12.4f}    "
              f"{result['val_metrics']['macro_f1']:>12.4f}    "
              f"{result['test_metrics']['micro_f1']:>12.4f}    "
              f"{result['test_metrics']['macro_f1']:>12.4f}    "
              f"{result['test_metrics']['hamming_loss']:>12.4f}")

print_comparison_table(stage1_results, "STAGE 1: MAIN CATEGORIES (6 labels)")
print_comparison_table(stage2_results, "STAGE 2: AI ETHICS ISSUES (12 labels)")

# ============================================================================
# SAVE RESULTS
# ============================================================================

print("\n" + "="*80)
print("SAVING RESULTS")
print("="*80)

# Save Stage 1 results
stage1_results_dict = {
    'stage': 'Stage 1 - Main Categories',
    'num_labels': len(stage1_labels),
    'labels': stage1_labels,
    'models': [
        {
            'name': r['model'],
            'training_time_minutes': r['training_time'] / 60,
            'val_metrics': r['val_metrics'],
            'test_metrics': r['test_metrics']
        }
        for r in stage1_results
    ]
}

with open('stage1_results.json', 'w') as f:
    json.dump(stage1_results_dict, f, indent=2)
print("✓ Stage 1 results saved to: stage1_results.json")

# Save Stage 2 results
stage2_results_dict = {
    'stage': 'Stage 2 - AI Ethics',
    'num_labels': len(stage2_labels),
    'labels': stage2_labels,
    'models': [
        {
            'name': r['model'],
            'training_time_minutes': r['training_time'] / 60,
            'val_metrics': r['val_metrics'],
            'test_metrics': r['test_metrics']
        }
        for r in stage2_results
    ]
}

with open('stage2_results.json', 'w') as f:
    json.dump(stage2_results_dict, f, indent=2)
print("✓ Stage 2 results saved to: stage2_results.json")

# ============================================================================
# VISUALIZATION
# ============================================================================

print("\n" + "="*80)
print("GENERATING VISUALIZATIONS")
print("="*80)

def plot_comparison(results, stage_name, filename):
    """Plot comparison charts"""
    models = [r['model'].split('/')[-1] if '/' in r['model'] else r['model'] for r in results]
    test_micro_f1 = [r['test_metrics']['micro_f1'] for r in results]
    test_macro_f1 = [r['test_metrics']['macro_f1'] for r in results]
    train_time = [r['training_time'] / 60 for r in results]
    
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    
    # Micro F1
    axes[0].bar(models, test_micro_f1, color='steelblue')
    axes[0].set_ylabel('Micro F1 Score')
    axes[0].set_title(f'{stage_name} - Test Micro F1')
    axes[0].set_ylim([0, 1])
    for i, v in enumerate(test_micro_f1):
        axes[0].text(i, v + 0.02, f'{v:.4f}', ha='center')
    
    # Macro F1
    axes[1].bar(models, test_macro_f1, color='coral')
    axes[1].set_ylabel('Macro F1 Score')
    axes[1].set_title(f'{stage_name} - Test Macro F1')
    axes[1].set_ylim([0, 1])
    for i, v in enumerate(test_macro_f1):
        axes[1].text(i, v + 0.02, f'{v:.4f}', ha='center')
    
    # Training Time
    axes[2].bar(models, train_time, color='lightgreen')
    axes[2].set_ylabel('Training Time (minutes)')
    axes[2].set_title(f'{stage_name} - Training Time')
    for i, v in enumerate(train_time):
        axes[2].text(i, v + max(train_time)*0.02, f'{v:.1f}m', ha='center')
    
    plt.tight_layout()
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"✓ Visualization saved: {filename}")
    plt.show()

plot_comparison(stage1_results, 'Stage 1', 'stage1_model_comparison.png')
plot_comparison(stage2_results, 'Stage 2', 'stage2_model_comparison.png')

# ============================================================================
# FINAL SUMMARY
# ============================================================================

print("\n" + "="*80)
print("TRAINING COMPLETE - FINAL SUMMARY")
print("="*80)

print("\n--- Best Models ---")

# Best Stage 1 model
best_s1 = max(stage1_results, key=lambda x: x['test_metrics']['micro_f1'])
print(f"\nStage 1 Best Model: {best_s1['model']}")
print(f"  Test Micro F1: {best_s1['test_metrics']['micro_f1']:.4f}")
print(f"  Test Macro F1: {best_s1['test_metrics']['macro_f1']:.4f}")

# Best Stage 2 model
best_s2 = max(stage2_results, key=lambda x: x['test_metrics']['micro_f1'])
print(f"\nStage 2 Best Model: {best_s2['model']}")
print(f"  Test Micro F1: {best_s2['test_metrics']['micro_f1']:.4f}")
print(f"  Test Macro F1: {best_s2['test_metrics']['macro_f1']:.4f}")

print("\n--- Next Steps ---")
print("1. Review the comparison tables and visualizations")
print("2. Select the best model for each stage")
print("3. Convert selected models to ONNX format")
print("4. Deploy to Flask app and Hugging Face Spaces")

print("\n" + "="*80)
print("✓ ALL DONE!")
print("="*80)
```

```python
# ============================================================================
# INSTALL REQUIRED PACKAGES
# ============================================================================

print("Installing required packages for ONNX conversion...")
import subprocess
import sys

# Install onnx and onnxruntime
packages_to_install = ['onnx', 'onnxruntime']

for package in packages_to_install:
    try:
        __import__(package)
        print(f"✓ {package} already installed")
    except ImportError:
        print(f"Installing {package}...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", package])
        print(f"✓ {package} installed")

# ============================================================================
# ONNX CONVERSION FOR BEST MODELS
# ============================================================================

import torch
import onnx
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import os
import json

# Try to import onnxruntime quantization (optional)
try:
    from onnxruntime.quantization import quantize_dynamic, QuantType
    QUANTIZATION_AVAILABLE = True
except ImportError:
    print("⚠ onnxruntime.quantization not available - will skip quantization")
    QUANTIZATION_AVAILABLE = False

print("\n" + "="*80)
print("ONNX MODEL CONVERSION")
print("="*80)

# ============================================================================
# CONFIGURATION
# ============================================================================

# Identify best models from results
best_stage1 = max(stage1_results, key=lambda x: x['test_metrics']['micro_f1'])
best_stage2 = max(stage2_results, key=lambda x: x['test_metrics']['micro_f1'])

print(f"\nBest Stage 1 Model: {best_stage1['model']}")
print(f"  Test Micro F1: {best_stage1['test_metrics']['micro_f1']:.4f}")

print(f"\nBest Stage 2 Model: {best_stage2['model']}")
print(f"  Test Micro F1: {best_stage2['test_metrics']['micro_f1']:.4f}")

# Output directories
ONNX_OUTPUT_DIR = './onnx_models'
os.makedirs(ONNX_OUTPUT_DIR, exist_ok=True)

# ============================================================================
# CONVERSION FUNCTION
# ============================================================================

def convert_to_onnx(model_path, output_name, num_labels, stage_name):
    """Convert a transformer model to ONNX format with quantization"""
    
    print(f"\n{'='*80}")
    print(f"Converting {stage_name}: {model_path}")
    print(f"{'='*80}")
    
    # Load model and tokenizer
    print("\n1. Loading model and tokenizer...")
    tokenizer = AutoTokenizer.from_pretrained(model_path)
    model = AutoModelForSequenceClassification.from_pretrained(model_path)
    model.eval()
    
    # Move model to CPU for ONNX export
    model = model.cpu()
    
    # Create dummy input
    print("2. Creating dummy input...")
    dummy_text = "This is a sample text for ONNX conversion."
    inputs = tokenizer(
        dummy_text,
        padding='max_length',
        max_length=MAX_LENGTH,
        truncation=True,
        return_tensors='pt'
    )
    
    # Prepare tuple of inputs for export
    dummy_input = (inputs['input_ids'], inputs['attention_mask'])
    
    # Define input/output names
    input_names = ['input_ids', 'attention_mask']
    output_names = ['logits']
    
    # Dynamic axes for variable batch size and sequence length
    dynamic_axes = {
        'input_ids': {0: 'batch_size', 1: 'sequence_length'},
        'attention_mask': {0: 'batch_size', 1: 'sequence_length'},
        'logits': {0: 'batch_size'}
    }
    
    # Export to ONNX
    onnx_path = f"{ONNX_OUTPUT_DIR}/{output_name}.onnx"
    print(f"3. Exporting to ONNX: {onnx_path}")
    
    try:
        with torch.no_grad():
            torch.onnx.export(
                model,
                dummy_input,
                onnx_path,
                input_names=input_names,
                output_names=output_names,
                dynamic_axes=dynamic_axes,
                opset_version=14,
                do_constant_folding=True,
                export_params=True,
                verbose=False
            )
        print("   ✓ ONNX export complete")
    except Exception as e:
        print(f"   ✗ ONNX export failed: {e}")
        return None
    
    # Verify ONNX model
    print("4. Verifying ONNX model...")
    try:
        onnx_model = onnx.load(onnx_path)
        onnx.checker.check_model(onnx_model)
        print("   ✓ ONNX model is valid")
    except Exception as e:
        print(f"   ⚠ ONNX verification warning: {e}")
    
    # Quantize model (INT8) for smaller size and faster inference
    quantized_path = onnx_path  # Default to non-quantized
    quantization_success = False
    
    if QUANTIZATION_AVAILABLE:
        print("5. Quantizing model (INT8)...")
        quantized_path = f"{ONNX_OUTPUT_DIR}/{output_name}_quantized.onnx"
        
        try:
            quantize_dynamic(
                onnx_path,
                quantized_path,
                weight_type=QuantType.QUInt8
            )
            print(f"   ✓ Quantized model saved: {quantized_path}")
            quantization_success = True
        except Exception as e:
            print(f"   ⚠ Quantization failed: {e}")
            print("   Using non-quantized model instead")
            quantized_path = onnx_path
    else:
        print("5. Skipping quantization (onnxruntime.quantization not available)")
    
    # Save tokenizer
    tokenizer_path = f"{ONNX_OUTPUT_DIR}/{output_name}_tokenizer"
    tokenizer.save_pretrained(tokenizer_path)
    print(f"6. Tokenizer saved: {tokenizer_path}")
    
    # Get file sizes
    original_size = os.path.getsize(onnx_path) / (1024**2)
    if quantization_success:
        quantized_size = os.path.getsize(quantized_path) / (1024**2)
    else:
        quantized_size = original_size
    
    print(f"\n--- Model Sizes ---")
    print(f"  Original ONNX: {original_size:.2f} MB")
    if quantization_success:
        print(f"  Quantized ONNX: {quantized_size:.2f} MB")
        print(f"  Size reduction: {((original_size - quantized_size) / original_size * 100):.1f}%")
    
    return {
        'onnx_path': onnx_path,
        'quantized_path': quantized_path,
        'tokenizer_path': tokenizer_path,
        'original_size_mb': original_size,
        'quantized_size_mb': quantized_size,
        'quantization_success': quantization_success
    }

# ============================================================================
# CONVERT STAGE 1 BEST MODEL
# ============================================================================

# Check if model has 'model_path' attribute (transformers only)
if 'model_path' in best_stage1:
    stage1_onnx = convert_to_onnx(
        best_stage1['model_path'],
        'stage1_best_model',
        len(stage1_labels),
        'Stage 1 (Main Categories)'
    )
else:
    print("\n⚠ Stage 1 best model is TF-IDF + LogReg (not transformer)")
    print("  ONNX conversion only applies to transformer models")
    print("  TF-IDF models will be saved separately using pickle")
    stage1_onnx = None
    
    # Save TF-IDF model if it's the best
    if best_stage1['model'] == 'TF-IDF + LogisticRegression':
        import pickle
        tfidf_dir = f"{ONNX_OUTPUT_DIR}/stage1_tfidf"
        os.makedirs(tfidf_dir, exist_ok=True)
        
        # Save vectorizer and model
        with open(f"{tfidf_dir}/vectorizer.pkl", 'wb') as f:
            pickle.dump(best_stage1['vectorizer'], f)
        with open(f"{tfidf_dir}/model.pkl", 'wb') as f:
            pickle.dump(best_stage1['model_obj'], f)
        
        # Save label information
        with open(f"{tfidf_dir}/labels.json", 'w') as f:
            json.dump(stage1_labels, f, indent=2)
        
        print(f"  ✓ TF-IDF model saved to: {tfidf_dir}")

# ============================================================================
# CONVERT STAGE 2 BEST MODEL
# ============================================================================

# Check if model has 'model_path' attribute (transformers only)
if 'model_path' in best_stage2:
    stage2_onnx = convert_to_onnx(
        best_stage2['model_path'],
        'stage2_best_model',
        len(stage2_labels),
        'Stage 2 (AI Ethics Issues)'
    )
else:
    print("\n⚠ Stage 2 best model is TF-IDF + LogReg (not transformer)")
    print("  ONNX conversion only applies to transformer models")
    print("  TF-IDF models will be saved separately using pickle")
    stage2_onnx = None
    
    # Save TF-IDF model if it's the best
    if best_stage2['model'] == 'TF-IDF + LogisticRegression':
        import pickle
        tfidf_dir = f"{ONNX_OUTPUT_DIR}/stage2_tfidf"
        os.makedirs(tfidf_dir, exist_ok=True)
        
        # Save vectorizer and model
        with open(f"{tfidf_dir}/vectorizer.pkl", 'wb') as f:
            pickle.dump(best_stage2['vectorizer'], f)
        with open(f"{tfidf_dir}/model.pkl", 'wb') as f:
            pickle.dump(best_stage2['model_obj'], f)
        
        # Save label information
        with open(f"{tfidf_dir}/labels.json", 'w') as f:
            json.dump(stage2_labels, f, indent=2)
        
        print(f"  ✓ TF-IDF model saved to: {tfidf_dir}")

# ============================================================================
# SAVE CONVERSION METADATA
# ============================================================================

print("\n" + "="*80)
print("SAVING CONVERSION METADATA")
print("="*80)

conversion_info = {
    'stage1': {
        'best_model': best_stage1['model'],
        'test_micro_f1': best_stage1['test_metrics']['micro_f1'],
        'test_macro_f1': best_stage1['test_metrics']['macro_f1'],
        'num_labels': len(stage1_labels),
        'labels': stage1_labels,
        'is_transformer': 'model_path' in best_stage1,
        'onnx_info': stage1_onnx
    },
    'stage2': {
        'best_model': best_stage2['model'],
        'test_micro_f1': best_stage2['test_metrics']['micro_f1'],
        'test_macro_f1': best_stage2['test_metrics']['macro_f1'],
        'num_labels': len(stage2_labels),
        'labels': stage2_labels,
        'is_transformer': 'model_path' in best_stage2,
        'onnx_info': stage2_onnx
    }
}

with open(f'{ONNX_OUTPUT_DIR}/conversion_info.json', 'w') as f:
    json.dump(conversion_info, f, indent=2)

print(f"✓ Conversion metadata saved: {ONNX_OUTPUT_DIR}/conversion_info.json")

# ============================================================================
# TEST ONNX INFERENCE (Optional but recommended)
# ============================================================================

print("\n" + "="*80)
print("TESTING ONNX INFERENCE")
print("="*80)

def test_onnx_inference(onnx_path, tokenizer_path, test_text):
    """Test ONNX model inference"""
    try:
        import onnxruntime as ort
    except ImportError:
        print("⚠ onnxruntime not available - skipping inference test")
        return None
    
    print(f"\nTesting: {onnx_path}")
    
    try:
        # Load tokenizer
        tokenizer = AutoTokenizer.from_pretrained(tokenizer_path)
        
        # Tokenize input
        inputs = tokenizer(
            test_text,
            padding='max_length',
            max_length=MAX_LENGTH,
            truncation=True,
            return_tensors='np'  # NumPy for ONNX Runtime
        )
        
        # Create ONNX Runtime session
        session = ort.InferenceSession(onnx_path)
        
        # Prepare inputs
        ort_inputs = {
            'input_ids': inputs['input_ids'].astype('int64'),
            'attention_mask': inputs['attention_mask'].astype('int64')
        }
        
        # Run inference
        outputs = session.run(None, ort_inputs)
        logits = outputs[0]
        
        # Apply sigmoid for multi-label classification
        import numpy as np
        probs = 1 / (1 + np.exp(-logits))  # Sigmoid
        predictions = (probs > 0.5).astype(int)
        
        print(f"  Input: {test_text[:50]}...")
        print(f"  Predictions shape: {predictions.shape}")
        print(f"  Predicted labels: {np.where(predictions[0] == 1)[0].tolist()}")
        print(f"  Probabilities (top 3): {np.sort(probs[0])[-3:][::-1]}")
        print("  ✓ Inference successful")
        
        return predictions
    except Exception as e:
        print(f"  ✗ Inference test failed: {e}")
        return None

# Test Stage 1 if available
if stage1_onnx:
    test_text = "Artificial intelligence is transforming the technology industry."
    test_onnx_inference(
        stage1_onnx['quantized_path'],
        stage1_onnx['tokenizer_path'],
        test_text
    )

# Test Stage 2 if available
if stage2_onnx:
    test_text = "The AI system raised concerns about bias and fairness in decision-making."
    test_onnx_inference(
        stage2_onnx['quantized_path'],
        stage2_onnx['tokenizer_path'],
        test_text
    )

# ============================================================================
# FINAL SUMMARY
# ============================================================================

print("\n" + "="*80)
print("ONNX CONVERSION COMPLETE")
print("="*80)

print("\n--- Converted Models ---")
if stage1_onnx:
    print(f"\nStage 1 (Transformer):")
    print(f"  ONNX Model: {stage1_onnx['onnx_path']}")
    print(f"  Quantized: {stage1_onnx['quantized_path']}")
    print(f"  Tokenizer: {stage1_onnx['tokenizer_path']}")
    print(f"  Size: {stage1_onnx['quantized_size_mb']:.2f} MB")
elif best_stage1['model'] == 'TF-IDF + LogisticRegression':
    print(f"\nStage 1 (TF-IDF):")
    print(f"  Model saved as pickle files in: {ONNX_OUTPUT_DIR}/stage1_tfidf/")

if stage2_onnx:
    print(f"\nStage 2 (Transformer):")
    print(f"  ONNX Model: {stage2_onnx['onnx_path']}")
    print(f"  Quantized: {stage2_onnx['quantized_path']}")
    print(f"  Tokenizer: {stage2_onnx['tokenizer_path']}")
    print(f"  Size: {stage2_onnx['quantized_size_mb']:.2f} MB")
elif best_stage2['model'] == 'TF-IDF + LogisticRegression':
    print(f"\nStage 2 (TF-IDF):")
    print(f"  Model saved as pickle files in: {ONNX_OUTPUT_DIR}/stage2_tfidf/")

print("\n--- Files to Download ---")
print(f"All models saved in: {ONNX_OUTPUT_DIR}/")
print("Download the entire directory for deployment")

print("\n--- Next Steps ---")
print("1. Download the ONNX models and tokenizers (or TF-IDF pickle files)")
print("2. Integrate into Flask app for inference")
print("3. Test with sample articles")
print("4. Deploy to Hugging Face Spaces")

print("\n✓ ALL DONE!")
print("="*80)
```

```python
# ============================================================================
# INFERENCE PIPELINE - BOTH STAGES
# ============================================================================

import numpy as np
import pickle
from transformers import AutoTokenizer
import torch

print("\n" + "="*80)
print("AI ETHICS RADAR - INFERENCE PIPELINE")
print("="*80)

# ============================================================================
# LOAD MODELS
# ============================================================================

print("\n--- Loading Stage 1 Model (Main Categories) ---")

# Stage 1: TF-IDF + LogReg
with open('./onnx_models/stage1_tfidf/vectorizer.pkl', 'rb') as f:
    stage1_vectorizer = pickle.load(f)
with open('./onnx_models/stage1_tfidf/model.pkl', 'rb') as f:
    stage1_model = pickle.load(f)
    
import json
with open('./onnx_models/stage1_tfidf/labels.json', 'r') as f:
    stage1_label_names = json.load(f)

# Clean up label names
stage1_labels_clean = [label.replace('stage1_binary_', '').replace('_', ' ').title() 
                       for label in stage1_label_names]

print(f"✓ Stage 1 Model loaded")
print(f"  Labels: {stage1_labels_clean}")

print("\n--- Loading Stage 2 Model (AI Ethics Issues) ---")

# Stage 2: RoBERTa ONNX
try:
    import onnxruntime as ort
    
    stage2_onnx_path = './onnx_models/stage2_best_model_quantized.onnx'
    stage2_tokenizer_path = './onnx_models/stage2_best_model_tokenizer'
    
    stage2_session = ort.InferenceSession(stage2_onnx_path)
    stage2_tokenizer = AutoTokenizer.from_pretrained(stage2_tokenizer_path)
    
    # Load label names from conversion info
    with open('./onnx_models/conversion_info.json', 'r') as f:
        conversion_info = json.load(f)
    stage2_label_names = conversion_info['stage2']['labels']
    
    # Clean up label names
    stage2_labels_clean = [label.replace('stage2_binary_', '').replace('_', ' ').title() 
                          for label in stage2_label_names]
    
    print(f"✓ Stage 2 Model loaded (ONNX)")
    print(f"  Labels: {stage2_labels_clean}")
    
    STAGE2_AVAILABLE = True
except Exception as e:
    print(f"⚠ Stage 2 model loading failed: {e}")
    STAGE2_AVAILABLE = False

# ============================================================================
# INFERENCE FUNCTIONS
# ============================================================================

def predict_stage1(text, threshold=0.5):
    """
    Predict main categories using TF-IDF + LogReg
    
    Args:
        text: Input article text
        threshold: Probability threshold (default 0.5)
    
    Returns:
        dict with predictions and probabilities
    """
    # Vectorize text
    X = stage1_vectorizer.transform([text])
    
    # Get predictions and probabilities
    predictions = stage1_model.predict(X)[0]
    probabilities = stage1_model.predict_proba(X)[0]
    
    # For multi-label, we need to extract probabilities for each label
    # OneVsRestClassifier returns list of probabilities for each classifier
    probs_per_label = []
    for i, estimator in enumerate(stage1_model.estimators_):
        prob = estimator.predict_proba(X)[0][1]  # Probability of class 1
        probs_per_label.append(prob)
    
    probs_per_label = np.array(probs_per_label)
    
    # Apply threshold
    predictions_threshold = (probs_per_label >= threshold).astype(int)
    
    # Get predicted labels
    predicted_indices = np.where(predictions_threshold == 1)[0]
    predicted_labels = [stage1_labels_clean[i] for i in predicted_indices]
    
    # Create results dictionary
    results = {
        'predictions': predictions_threshold,
        'probabilities': probs_per_label,
        'predicted_labels': predicted_labels,
        'all_labels_with_probs': [
            {'label': stage1_labels_clean[i], 'probability': float(probs_per_label[i])}
            for i in range(len(stage1_labels_clean))
        ]
    }
    
    return results

def predict_stage2(text, threshold=0.5):
    """
    Predict AI ethics issues using RoBERTa ONNX
    
    Args:
        text: Input article text
        threshold: Probability threshold (default 0.5)
    
    Returns:
        dict with predictions and probabilities
    """
    if not STAGE2_AVAILABLE:
        return None
    
    # Tokenize input
    inputs = stage2_tokenizer(
        text,
        padding='max_length',
        max_length=MAX_LENGTH,
        truncation=True,
        return_tensors='np'
    )
    
    # Prepare ONNX inputs
    ort_inputs = {
        'input_ids': inputs['input_ids'].astype('int64'),
        'attention_mask': inputs['attention_mask'].astype('int64')
    }
    
    # Run inference
    outputs = stage2_session.run(None, ort_inputs)
    logits = outputs[0][0]
    
    # Apply sigmoid for multi-label classification
    probabilities = 1 / (1 + np.exp(-logits))
    
    # Apply threshold
    predictions = (probabilities >= threshold).astype(int)
    
    # Get predicted labels
    predicted_indices = np.where(predictions == 1)[0]
    predicted_labels = [stage2_labels_clean[i] for i in predicted_indices]
    
    # Create results dictionary
    results = {
        'predictions': predictions,
        'probabilities': probabilities,
        'predicted_labels': predicted_labels,
        'all_labels_with_probs': [
            {'label': stage2_labels_clean[i], 'probability': float(probabilities[i])}
            for i in range(len(stage2_labels_clean))
        ]
    }
    
    return results

def predict_article(text, threshold=0.5, verbose=True):
    """
    Complete inference pipeline for both stages
    
    Args:
        text: Input article text
        threshold: Probability threshold (default 0.5)
        verbose: Print detailed results
    
    Returns:
        dict with both stage results
    """
    if verbose:
        print("\n" + "="*80)
        print("ARTICLE CLASSIFICATION")
        print("="*80)
        print(f"\nArticle Preview (first 200 chars):")
        print(f"{text[:200]}...")
        print(f"\nThreshold: {threshold}")
    
    # Stage 1: Main Categories
    if verbose:
        print("\n" + "-"*80)
        print("STAGE 1: MAIN CATEGORIES")
        print("-"*80)
    
    stage1_results = predict_stage1(text, threshold)
    
    if verbose:
        if stage1_results['predicted_labels']:
            print(f"\n✓ Predicted Categories ({len(stage1_results['predicted_labels'])}):")
            for label in stage1_results['predicted_labels']:
                idx = stage1_labels_clean.index(label)
                prob = stage1_results['probabilities'][idx]
                print(f"  • {label}: {prob:.4f} ({prob*100:.2f}%)")
        else:
            print("\n⚠ No categories predicted above threshold")
        
        print(f"\nAll Categories (sorted by probability):")
        sorted_labels = sorted(stage1_results['all_labels_with_probs'], 
                              key=lambda x: x['probability'], reverse=True)
        for item in sorted_labels:
            marker = "✓" if item['probability'] >= threshold else " "
            print(f"  {marker} {item['label']:<30} {item['probability']:.4f} ({item['probability']*100:.2f}%)")
    
    # Stage 2: AI Ethics Issues
    if verbose:
        print("\n" + "-"*80)
        print("STAGE 2: AI ETHICS ISSUES")
        print("-"*80)
    
    if STAGE2_AVAILABLE:
        stage2_results = predict_stage2(text, threshold)
        
        if verbose:
            if stage2_results['predicted_labels']:
                print(f"\n✓ Predicted Ethics Issues ({len(stage2_results['predicted_labels'])}):")
                for label in stage2_results['predicted_labels']:
                    idx = stage2_labels_clean.index(label)
                    prob = stage2_results['probabilities'][idx]
                    print(f"  • {label}: {prob:.4f} ({prob*100:.2f}%)")
            else:
                print("\n⚠ No ethics issues predicted above threshold")
            
            print(f"\nAll Ethics Issues (sorted by probability):")
            sorted_labels = sorted(stage2_results['all_labels_with_probs'], 
                                  key=lambda x: x['probability'], reverse=True)
            for item in sorted_labels:
                marker = "✓" if item['probability'] >= threshold else " "
                print(f"  {marker} {item['label']:<30} {item['probability']:.4f} ({item['probability']*100:.2f}%)")
    else:
        stage2_results = None
        if verbose:
            print("\n⚠ Stage 2 model not available")
    
    if verbose:
        print("\n" + "="*80)
    
    return {
        'stage1': stage1_results,
        'stage2': stage2_results,
        'text_preview': text
    }

# ============================================================================
# TEST EXAMPLES
# ============================================================================

print("\n" + "="*80)
print("TESTING INFERENCE PIPELINE")
print("="*80)

# Example 1: AI Technology Article
example1 = """
Google's new artificial intelligence system has achieved breakthrough performance in natural language understanding. 
The deep learning model, trained on massive datasets, can now generate human-like text and answer complex questions. 
However, researchers warn about potential biases in the training data that could lead to unfair outcomes. 
The technology industry is watching closely as this AI advancement could transform software development and business operations.
"""

print("\n" + "="*80)
print("EXAMPLE 1: AI Technology Article")
print("="*80)
results1 = predict_article(example1, threshold=0.5)

# Example 2: Sports Article
example2 = """
The Los Angeles Lakers secured a dramatic victory in last night's basketball game. 
LeBron James scored 35 points, leading his team to a 112-108 win over the Boston Celtics. 
The game went into overtime, with both teams displaying exceptional athleticism and teamwork. 
Fans filled the arena, creating an electric atmosphere as the final buzzer sounded.
"""

print("\n" + "="*80)
print("EXAMPLE 2: Sports Article")
print("="*80)
results2 = predict_article(example2, threshold=0.5)

# Example 3: AI Ethics Article
example3 = """
A new study reveals significant gender and racial bias in facial recognition systems used by law enforcement. 
The AI algorithms showed error rates up to 34% higher for women of color compared to white men. 
Privacy advocates are calling for stricter regulations and transparency requirements for AI systems. 
The findings raise serious concerns about algorithmic fairness and accountability in automated decision-making. 
Experts emphasize the need for diverse training data and rigorous testing to prevent discriminatory outcomes.
"""

print("\n" + "="*80)
print("EXAMPLE 3: AI Ethics Article")
print("="*80)
results3 = predict_article(example3, threshold=0.9)

# ============================================================================
# INTERACTIVE INFERENCE
# ============================================================================

print("\n" + "="*80)
print("INTERACTIVE INFERENCE")
print("="*80)
print("\nNow you can classify your own articles!")
print("Copy the code below and modify 'your_article' variable:\n")

print("="*80)
print("# YOUR ARTICLE HERE")
print("="*80)

your_article = '''
The Shifting Dynamics of Modern Politics

Modern politics is shaped by evolving public expectations, shifting party alliances, and the growing importance of effective leadership. Citizens today are more engaged and vocal, demanding transparency, accountability, and policies that reflect their daily realities. As a result, political parties are reassessing their platforms to appeal to broader and more diverse populations.

Elections have become increasingly competitive, with candidates focusing on issues such as governance, economic stability, and national security. Debates and public forums play a crucial role in helping voters evaluate political agendas and leadership qualities. Campaign strategies now emphasize direct communication with constituents, allowing politicians to address concerns and present their policy priorities more clearly.

International relations also influence domestic political landscapes. Governments must navigate diplomatic challenges, form alliances, and respond to global developments that impact national interests. Political leaders are expected to balance these external pressures while crafting legislation that supports growth and stability at home.

Public participation remains a cornerstone of political systems. Voter turnout, civic engagement, and local activism shape the direction of policy and governance. As societies continue to change, political institutions face ongoing pressure to adapt and respond effectively to the needs of the people.
'''

# Run inference with default threshold (0.5)
results = predict_article(your_article, threshold=0.9)

# Or try different threshold (e.g., 0.3 for more predictions, 0.7 for more confident)
results = predict_article(your_article, threshold=0.9)

print("\n" + "="*80)
print("✓ INFERENCE PIPELINE READY!")
print("="*80)
print("\nUsage:")
print("  results = predict_article(your_text, threshold=0.5)")
print("  results = predict_article(your_text, threshold=0.3, verbose=False)  # Less output")
print("\n" + "="*80)
```
